```python
import numpy as np
from sklearn.cluster import KMeans
import scipy.misc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = scipy.misc.face()
```

```python
plt.imshow(f)
```

```python
f1=scipy.misc.face(gray=True)
```

```python
plt.figure(figsize=(10, 4))
plt.imshow(f1, cmap=plt.cm.gray)
plt.show()
```

```python
image = f1.reshape(f1.shape[0]*f1.shape[1],1)
kmeans = KMeans(n_clusters = 5)
kmeans.fit(image)
```

```python
kmeans.cluster_centers_
```

```python
len(kmeans.labels_)
```

```python
labels = np.asarray(kmeans.labels_)
```

```python
labels.reshape(f1.shape[0],f1.shape[1])
```

```python
clusters = np.asarray(kmeans.cluster_centers_) 
labels = np.asarray(kmeans.labels_)  
labels = labels.reshape(f1.shape[0],f1.shape[1]); 
```

```python
plt.imsave('compressed_racoon_image.png',labels);
```

```python
plt.imshow(f1, cmap=plt.cm.gray)
plt.show()
plt.imshow(labels)
plt.show()
```
